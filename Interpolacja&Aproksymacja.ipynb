```python
from matplotlib import axes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
from scipy import interpolate
from scipy.fft import fft
from scipy.optimize import curve_fit  
import math
```

```python
def Lagrange_Interpolation(x, y, xx):

  n = len(x)
  sum = 0
  for i in range(n):
    product = y[i]
    for j in range(n):
      if i!= j:
        product = product*(xx - x[j])/(x[i] - x[j])
    sum = sum + product
  return sum
```

```python
#Wykres
df1 = pd.read_csv(r'/content/^oseax_m (1).csv', sep=";")
np.array
d1 = np.array(df1["Data"])
z1 = np.array(df1["Zamkniecie"])

fig,axes=plt.subplots(1,1)
axes.plot(d1, z1, '.')
plt.grid()
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Wykres OSEAX ')

for i, tick in enumerate(axes.xaxis.get_ticklabels()):
    if i % 110 != 0:
        tick.set_visible(False)
#Lagrange
x1 = np.linspace(1, df1.shape[0], df1.shape[0])
xx = np.linspace(1, df1.shape[0], df1.shape[0])
yy1 = [Lagrange_Interpolation(x1, z1, i) for i in xx]

fig,axes=plt.subplots(1,1)
axes.plot(d1, yy1, 'g')
plt.grid()
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Lagrange OSEAX')

for i, tick in enumerate(axes.xaxis.get_ticklabels()):
    if i % 110 != 0:
        tick.set_visible(False)
#interp1d
f1 = sc.interpolate.interp1d(x1, z1)

ynew1 = f1(x1)

fig,axes=plt.subplots(1,1)
axes.plot(d1, ynew1, 'r')
plt.grid()
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Interp1d OSEAX')

for i, tick in enumerate(axes.xaxis.get_ticklabels()):
    if i % 110 != 0:
        tick.set_visible(False)
#UnivariateSpline
f1 = sc.interpolate.UnivariateSpline(x1, z1)

new1 = f1(x1)

fig,axes=plt.subplots(1,1)
axes.plot(d1, new1, 'y')
plt.grid()
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('UnivariateSpline OSEAX')

for i, tick in enumerate(axes.xaxis.get_ticklabels()):
    if i % 110 != 0:
        tick.set_visible(False)
#Akima1DInterpolator
f1 = sc.interpolate.Akima1DInterpolator(x1, z1)

new1 = f1(x1)

fig,axes=plt.subplots(1,1)
axes.plot(d1, new1, 'k')
plt.grid()
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Akima1DInterpolator OSEAX')

for i, tick in enumerate(axes.xaxis.get_ticklabels()):
    if i % 110 != 0:
        tick.set_visible(False)
```

```python
def TransformataFouriera(t, y):
  N = len(t)
  dt = t[1] - t[0]
  yf = 2.0 / N *np.abs(fft(y)[0:N//2])
  xf = np.fft.fftfreq(N, d=dt)[0:N //2]
  return xf, yf
```

```python
#OSEAX
xf, yf = TransformataFouriera(x1, z1)
plt.plot(xf, yf)
plt.grid()
plt.show()
```
